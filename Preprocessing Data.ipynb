```python
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Take a column of data and normalize it
# The array is playing the part of the values in the column

# If a feature has a variance that is orders of magnitude larger 
# than others, it might dominate the objective function and make 
# the estimator unable to learn from other features correctly as expected.

# The function scale provides a quick and easy way to perform this 
# operation on a single array-like dataset:
```

```python
# Create demo array

X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Plot array without normalization

ax1 = plt.plot(X_train)

plt.show(ax1)
```

```python
# Let's scale the array

X_scaled = preprocessing.scale(X_train)

X_scaled
```

```python
# Scaled data has zero mean

X_scaled.mean(axis = 0)
```

```python
# Scaled data has unit variance

X_scaled.std(axis = 0)
```

```python
# Plot array with normalization

# Notice the 0 is the middle integer on the right now
# That's the 0 mean
# Also notice how the shape has remained pretty much the same

ax = plt.plot(X_scaled)

plt.show(ax)
```
